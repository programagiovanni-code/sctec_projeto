# Tratando nulos e duplicados

```python
# Confirmando os id repetidos - Para ter certeza
df_id = df_original[df_original.duplicated(subset=['id'], keep=False)] 
df_id.sort_values(by='id')
```

```python
# Visualizando e entendendo os nulos
df_nulos = df_original[df_original.isna().any(axis=1)]
df_nulos
```

```python
# Ver se não se repete pelo id --> (Não deu)
display(df_original[df_original['id']==1736800520])
display(df_original[df_original['id']==6865200140])
```

```python
# Confirmando hipotese de area construida ser porão + geral
df_area = df_original[['sqft_living','sqft_above','sqft_basement']]
df_area['hipotese_soma'] = df_area['sqft_above'] + df_area['sqft_basement']

diferencas = df_area[df_area['hipotese_soma'] != df_area['sqft_living']]
diferencas
```

```python
# Substituindo os nulos pelo calculo (living - basement)
df_semnulos = df_original.copy()
df_semnulos['sqft_above'] = df_semnulos['sqft_living'] - df_semnulos['sqft_basement'] 
```
